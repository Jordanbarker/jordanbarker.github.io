```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

plt.rcParams['figure.figsize'] = (12, 4)
rng = np.random.default_rng(42)

x = np.linspace(0, 10, 100)
data = {'x': x}
for i in range(10):
    data[i] = np.sin(x) + rng.normal(1, 0.1, len(x))
    plt.plot(x, data[i])
    data[i+10] = np.cos(x) + rng.normal(1, 0.1, len(x))
    plt.plot(x, data[i+10])

df = pd.DataFrame(data)
plt.title('Multiple Time Series')
plt.ylabel('y')
plt.xlabel('x')
plt.legend()
plt.show()
```
